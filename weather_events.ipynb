```python
import pandas as pd

weather_event_data = "Resources/WeatherEvents_Jan2016-Dec2021.csv"

weather_event_df = pd.read_csv(weather_event_data)

weather_event_df.head()
```

```python
# parsing out dates
new = weather_event_df["StartTime(UTC)"].str.split("-", n = 2, expand = True)
# new column for year from start of weather event
weather_event_df["year"]= new[0]
# new column for month
weather_event_df["month"]= new[1]
#parse out portion of date not needed.
weather_event_df["junk"]= new[2]
# drop junk.
weather_event_df.drop(columns =["junk"], inplace = True)
# did it work? yes.
weather_event_df.head()
```

```python
# dropping columns we don't need
master_df = weather_event_df[['Type','Severity','Precipitation(in)','AirportCode','LocationLat','LocationLng','City','County','State','ZipCode','year','month']]
master_df = pd.DataFrame(master_df)
master_df
```

```python
# dataframe of weather types
Type_df = weather_event_df.drop_duplicates(["Type"]).reset_index(drop=True)[["Type"]]
Type_df = pd.DataFrame(Type_df.rename(columns={'Type':'Weather_Type'}))
Type_df
```

```python
# Datframe with airport code and corresponding location data
location_df = weather_event_df.drop_duplicates(["AirportCode", "LocationLat", "LocationLng", "City", "State", "ZipCode"]).reset_index(drop=True)[["AirportCode", "LocationLat", "LocationLng", "City", "State", "ZipCode"]]
location_df = pd.DataFrame(location_df)
location_df
```

```python
#Dataframe of severity types
severity_df = weather_event_df.drop_duplicates(["Severity"]).reset_index(drop=True)[["Severity"]]
severity_df = pd.DataFrame(severity_df)
severity_df
```

```python
# Dataframe consisting of Weather typr, Severity, Precip, year, and month
weather_df = weather_event_df[["AirportCode","Type", "Severity", "Precipitation(in)", "year", "month"]]
weather_df = pd.DataFrame(weather_df)
weather_df
```
